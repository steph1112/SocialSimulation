```python
import itertools
from random import choice
import pandas as pd

# Ejemplo de sociedad y estrategias
society = [{'id': 1, 'name': 'Alice', 'strategy': None, 'score': 0},
           {'id': 2, 'name': 'Bob', 'strategy': None, 'score': 0},
           {'id': 3, 'name': 'Jane', 'strategy': None, 'score': 0},
           {'id': 4, 'name': 'Zoe', 'strategy': None, 'score': 0},
          {'id': 4, 'name': 'Ricky', 'strategy': None, 'score': 0},
          {'id': 4, 'name': 'Gonzalo', 'strategy': None, 'score': 0},
          {'id': 4, 'name': 'Nicole', 'strategy': None, 'score': 0},
          {'id': 4, 'name': 'Luis', 'strategy': None, 'score': 0},
          {'id': 4, 'name': 'Alejandro', 'strategy': None, 'score': 0},
          {'id': 4, 'name': 'Karla', 'strategy': None, 'score': 0}]
    

strategies = ['Piedra', 'Papel', 'Tijeras', 'Lagarto', 'Spock']

def payoff(strategy1, strategy2):
    if strategy1 == strategy2:
        return (0.5, 0.5)
    elif (strategy1 == 'Piedra' and strategy2 in ['Tijeras', 'Lagarto']) or \
         (strategy1 == 'Papel' and strategy2 in ['Piedra', 'Spock']) or \
         (strategy1 == 'Tijeras' and strategy2 in ['Papel', 'Lagarto']) or \
         (strategy1 == 'Lagarto' and strategy2 in ['Spock', 'Papel']) or \
         (strategy1 == 'Spock' and strategy2 in ['Tijeras', 'Piedra']):
        return (1, 0)
    else:
        return (0, 1)

for aRound in range(100):  # esto sucederá 100 veces, todos juegan contra todos
    # en cada ronda:
    for player1, player2 in itertools.combinations(society, 2):
        # cada uno elige estrategia
        player1['strategy'] = choice(strategies)
        player2['strategy'] = choice(strategies)
        
        # resultado de las estrategias elegidas
        print(player1['name'] + " chose " + player1['strategy'] + " - " + player2['name'] + " chose " + player2['strategy'])
        result = payoff(player1['strategy'], player2['strategy'])
        if result == (1, 0):
            print(player1['name'] + " won")
            player1['score'] += 1
        elif result == (0, 1):
            print(player2['name'] + " won")
            player2['score'] += 1
        else:
            print("TIE!")
            player1['score'] += 0.5
            player2['score'] += 0.5
```
