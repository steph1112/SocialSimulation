```python
import itertools
from random import choice
import pandas as pd

# Ejemplo de sociedad y estrategias
society = [{'id': 1, 'name': 'Jim', 'strategy': None, 'score': 0},
           {'id': 2, 'name': 'Jane', 'strategy': None, 'score': 0},
           {'id': 3, 'name': 'Peter', 'strategy': None, 'score': 0},
           {'id': 4, 'name': 'Zoe', 'strategy': None, 'score': 0}]

strategies = ['Rock', 'Paper', 'Scissors']

def payoff(strategy1, strategy2):
    if strategy1 == strategy2:
        return (0.5, 0.5)
    elif (strategy1 == 'Rock' and strategy2 == 'Scissors') or \
         (strategy1 == 'Scissors' and strategy2 == 'Paper') or \
         (strategy1 == 'Paper' and strategy2 == 'Rock'):
        return (1, 0)
    else:
        return (0, 1)

for aRound in range(20):  # esto sucederá 1000 veces, todos juegan contra todos
    # en cada ronda:
    for player1, player2 in itertools.combinations(society, 2):
        # cada uno elige estrategia
        player1['strategy'] = choice(strategies)
        player2['strategy'] = choice(strategies)
        
        # resultado de las estrategias elegidas
        print(player1['name'] + " chose " + player1['strategy'] + " - " + player2['name'] + " chose " + player2['strategy'])
        result = payoff(player1['strategy'], player2['strategy'])
        if result == (1, 0):
            print(player1['name'] + " won")
            player1['score'] += 1
        elif result == (0, 1):
            print(player2['name'] + " won")
            player2['score'] += 1
        else:
            print("TIE!")
            player1['score'] += 0.5
            player2['score'] += 0.5
```
